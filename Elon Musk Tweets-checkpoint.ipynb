```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import matplotlib.pyplot as plt
import re
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()


import matplotlib.pyplot as plt; plt.rcdefaults()
```

```python
# list to append tweet data
tweets_list = []

# using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:elonmusk').get_items()):
    if i>10000:
        break
    tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above 
df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
starting_date = df["Datetime"].min()
ending_date = df["Datetime"].max()

print("Starting date: ", starting_date)
print("Ending date: ", ending_date)
```

```python
# cleaning tweet text:

def cleaner(tweet):
    tweet = re.sub("["u"\U0001F600-\U0001F64F"u"\U0001F300-\U0001F5FF"u"\U0001F680-\U0001F6FF"u"\U0001F1E0-\U0001F1FF""]+", "", tweet) #deleting (most) emoticons
    tweet = re.sub("@[A-Za-z0-9]+","",tweet) #deleting username references
    tweet = re.sub("[0-9]+","",tweet) #deleting numbers
    tweet = re.sub(r"http\S+","",tweet) 
    tweet = re.sub(r"https\S+","",tweet) #deleting URLs
    tweet = " ".join(tweet.split())
    tweet = tweet.replace("#", "").replace("_", " ")
    return tweet
df["clean"] = df["Text"].map(lambda x: cleaner(x))
```

```python
# extracting tokens
tokenizer = nlp.tokenizer
tokens = tokenizer("".join(str(df.clean.tolist())))
print(tokens)
```

```python
doc = nlp(tokens)
items = [x.text for x in tokens.ents]
word_counts = Counter(items).most_common(25)
```

```python
df_text = pd.DataFrame(word_counts, columns=["text", "count"])
df_text.head()
```

```python
#creating a graph of Elon's most popular twitter words
plt.rcParams.update({'font.size': 7})
fig, ax = plt.subplots()

ax.barh(df_text["text"], df_text["count"], align='center', color='red')
ax.invert_yaxis()
ax.set_title("Elon Musk's Most Popular Twitter Words 04/05/2019 to 05/18/2022")
ax.set_ylabel('Word from Tweet')
ax.set_xlabel('Count of Words')

plt.show()
```

```python
#grouping word frequencies by every 200 tweets
my_list =[]
for i in range(1, 51):
    for j in range(200):
        my_list.append(i)

my_list.append(51)
df["Tweet Tracker"] = my_list
df["tesla_check"] = df["clean"].str.contains("Tesla")
df["mars_check"] = df["clean"].str.contains("Mars")
df["twitter_check"] = df["clean"].str.contains("Twitter")
```

```python
day = df.iloc[::200, :]
every_twohundred_tweets = list(day["Tweet Tracker"])
every_twohundred_tweets.reverse()
```

```python
# creating a new data frame with word frequences per 200 tweets
tesla = list(df.groupby("Tweet Tracker")["tesla_check"].sum())
mars = list(df.groupby("Tweet Tracker")["mars_check"].sum())
twitter = list(df.groupby("Tweet Tracker")["twitter_check"].sum())

frequencies = pd.DataFrame(every_twohundred_tweets, columns=['Day Tracker'])
frequencies["tesla"] = tesla
frequencies["mars"] = mars
frequencies["twitter"] = twitter
frequencies.head()
```

```python
fig, ax = plt.subplots()
ax.plot(frequencies["Day Tracker"], frequencies["tesla"], linewidth=1.0, alpha=0.5, color="red")
ax.plot(frequencies["Day Tracker"], frequencies["mars"], linewidth=1.0, alpha=0.5, color="blue")
ax.plot(frequencies["Day Tracker"], frequencies["twitter"], linewidth=1.0, alpha=0.5, color="green")
ax.legend(['"Tesla"', '"Mars"', '"Twitter"'])
ax.set_title("Elon Musk Twitter Word Frequencies May 2019 - May 2022")
ax.set_ylabel("# of Uses per 200 Tweets")
ax.set_xlabel("x200 Tweets")
ax.grid()
```

```python
# Key points:
# Elon tweets "Twitter" increasingly more often beginning around April when he announces he will be buying Twitter (source: https://www.theverge.com/23026874/elon-musk-twitter-buyout-news-updates)
# Elon's frequency of tweeting "Tesla" spikes around November 2021 (around his 8,500th tweet since 02/09/2019) when he sells $5.7 billion worth of shares (source: https://www.cnbc.com/2021/11/12/tesla-stock-had-its-worst-week-in-20-months-after-musk-sold-shares.html)
```

```python
df["tesla_cumsum"] = df["tesla_check"].cumsum()
df["mars_cumsum"] = df["mars_check"].cumsum()
df["twitter_cumsum"] = df["twitter_check"].cumsum()
df
```

```python
datetimereversed = list(df["Datetime"])
datetimereversed.reverse()
df["Datetimereversed"] = datetimereversed
```

```python
fig, ax = plt.subplots()

ax.plot(df["Datetimereversed"], df["tesla_cumsum"], linewidth=1.0, alpha=0.5, color="red")
ax.plot(df["Datetimereversed"], df["mars_cumsum"], linewidth=1.0, alpha=0.5, color="blue")
ax.plot(df["Datetimereversed"], df["twitter_cumsum"], linewidth=1.0, alpha=0.5, color="green")
ax.legend(["Tesla", "Mars", "Twitter"])
ax.set_title("Elon Musk Twitter Word Frequencies May 2019 - May 2022")
ax.set_ylabel("# of Uses per 200 Tweets")
ax.set_xlabel("Date")
ax.grid()
```
